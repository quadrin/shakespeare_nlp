```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.linear_model import LogisticRegression

text_clf = Pipeline([('vect', CountVectorizer(ngram_range=(1,5))),
                     ('tfidf', TfidfTransformer()),
                     ('clf', LogisticRegression(penalty = 'l2', max_iter=30, random_state=10, class_weight = 'balanced'))

])
```

```python
text_clf.fit(poems, poet_order)
```

```python
predictions = text_clf.predict_proba(poems)
```

```python
text_clf.classes_
```

```python
import pandas as pd
```

```python
pred_df = pd.DataFrame(predictions, columns=text_clf.classes_)
pred_df["true_value"] = poet_order
pred_df.head()
```

```python
pred_df.groupby("true_value").mean()
```

```python
sklearn.metrics.confusion_matrix(poet_order, text_clf.predict(poems), labels=poets, sample_weight=None, normalize=None)
```
